```python
# homework 1 chapter 7
from matplotlib import pyplot as plt
from scipy.io import wavfile
import numpy as np

# read file
samplerate, samples = wavfile.read('yamaha.wav')
# get the time of each sample
t = np.arange(len(samples))/float(samplerate)

# plot the spectogram
plt.figure(figsize=(50, 10))
plt.fill_between(t, samples[:,0], samples[:,1], color='b') 
plt.xlim(t[0], t[-1])
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```

```python
# homework 1 chapter 8
# Python 3.6 could not read url therefore I put the text into an .txt file

import numpy as np
import pandas
from collections import Counter

# read text from file

text = open("text.txt", "r")
text = text.read().splitlines()

# convert to numpy
text = np.asarray(text)

# split the text to character and append it to a new numpy array, and print it out
text_splitted = np.char.split(text)
print(text_splitted)
print('\n')
block = np.array([])
for wordy in text_splitted:
    for word in wordy:
        for w in word:
            block = np.append(block, w)
print (block)

# count the elements and plot the histogram
block.transpose()
letter_counts = Counter(block)
datas = pandas.DataFrame.from_dict(letter_counts, orient = 'index')
datas.plot(kind = 'bar')

```
